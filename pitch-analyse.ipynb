```python
import numpy as np
import matplotlib.pyplot as plt
import os
import math
```

```python
BASE_DIR = "./poses"
scenes_list = os.listdir(BASE_DIR)
```

```python
def Rotation_to_Euler(R):
    yaw = np.arctan2(R[2][0],R[2][1])
    pitch = np.arccos(R[2][2])
    roll = -np.arctan2(R[0][2],R[1][2])
    return yaw,pitch,roll
```

```python
global_pitch = []
```

```python
for scene in scenes_list:
    poses_list = os.listdir(os.path.join(BASE_DIR,scene))
    yaw = []
    pitch = []
    roll = []
    for pose in poses_list:
        R = np.loadtxt(os.path.join(BASE_DIR,scene,pose))[:3,:3]
        if(-float('inf') in R):
            continue
        y,p,r = Rotation_to_Euler(R)
        yaw.append(y*180/math.pi)
        pitch.append(p*180/math.pi)
        roll.append(r*180/math.pi)
        global_pitch.append(p*180/math.pi)

    percentile_range = np.linspace(0,100,50)
    output = np.percentile(pitch,percentile_range)
    plt.plot(percentile_range,output)
    plt.xlabel('percentile')
    plt.ylabel('Pitch in Degree')
    plt.savefig('./pitch_plots/'+scene+'.png')
    plt.close()
```
